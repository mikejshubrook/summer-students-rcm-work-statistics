```python
import numpy as np
import pandas as pd
import os
import sys
import matplotlib.pyplot as plt
```

```python
# test script to import and plot dynamics

# parameters to use

### Numerical parameters for simulation ###
T0=0.0 # initial time
T1=10.0 # final time
DT=0.1 # time step


### Bath parameters ###
TEMPERATURE=300.0 # temperature [K]
GAMMA=0.001 # spectral density width
W0=0.05 # spectral density peak location
ALPHA=0.1 # coupling strength

UNITS='eV'

### TLS parameters ###
EPSILON=2.0 # energy splitting of the TLS
DELTA=0.1 # driving amplitude
OMEGA=0.0 # driving frequency
CONV=1519.3 # conversion factor

### RC parameters
M=10 # dimension of RC Hilbert space

# create filename
filename = f"dynamics-files/dynamics_{UNITS}_A{ALPHA}_E{EPSILON}_D{DELTA}_O{OMEGA}_T{TEMPERATURE}_G{GAMMA}_w0{W0}_t0{T0}_tf{T1}_dt{DT}_M{M}.csv"
# check if file exists
if not os.path.exists(filename):
    print(f"File {filename} does not exist.")
    sys.exit(1)

# read data from file
df = pd.read_csv(filename)

df.head()
```

```python
# plot the different quantities

fig, axs = plt.subplots(1, 3, figsize=(20, 4))

# Plot excited population
axs[0].plot(df['tlist'], df['excited_pop'], label='Excited Population', color='blue')
axs[0].set_title('Excited Population Over Time')
axs[0].set_xlabel('Time (ps)')
axs[0].set_ylabel('Excited Population')
axs[0].grid()   
axs[0].legend()

# Plot coherence
axs[1].plot(df['tlist'], df['coherence'], label='Coherence', color='orange')
axs[1].set_title('Coherence Over Time')
axs[1].set_xlabel('Time (ps)')
axs[1].set_ylabel('Coherence')
axs[1].grid()       
axs[1].legend()



# Plot RC population    
axs[2].plot(df['tlist'], df['rc_pop'], label='RC Population', color='green')
axs[2].set_title('RC Population Over Time')
axs[2].set_xlabel('Time (ps)')
axs[2].set_ylabel('RC Population')
axs[2].grid()
axs[2].legend()
```
